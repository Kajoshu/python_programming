## Question4 

```python
#Program to print the factorial of a given number.
number = int(input("Enter a number"))
factorial = 1
for i in range(0,number):
    factorial *= (i+1)
print(factorial)
```

## Question7

```python
#Program to print if a number is a prime number or not
number1 = int(input("Enter a number:"))
primer = 0
for i in range(2,number1):
    if number1 % i == 0:
        primer = primer + 1    
    else:
        primer = primer
if primer == 0:
    prime = "It is a prime number"
else:
    prime = "It is not a prime number"
    
print(prime)
```
